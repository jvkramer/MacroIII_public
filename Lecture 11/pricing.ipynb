```python
#### Simulate a model with Fischer, Taylor and Calvo contracts

import numpy as np
import matplotlib.pyplot as plt

CB =  ['#000000','#004488','#BB5566','#DDAA33', 'purple']
LS = ['solid', 'dashed', ':', '-.']

def plot_phil(,,,npoints):
    
    ## Drawing the New Keynesian Phillips Curve
    y_t = np.linspace(-1,1,1000)
    err = np.random.normal(0,0.1,1000) ##error in expectation
    
    ## Model future expectation as current inflation plus some error
    #  (t) =  y(t) +  E[(t+1)] 
    #       =  y(t) +  (t)+ e 
    # (1-) (t)    =  y(t) + e               
     = *(1-*(1-))*(-1)/(1-)
    pi_t = /(1-) * y_t + err/(1-)
    pi_exact=  /(1-) * y_t

    plt.scatter(y_t,pi_t)
    plt.plot(y_t,pi_exact, color = 'red',linewidth=4.0)
    plt.annotate(r'$ \phi = $' + str(),(-0.75,7.5))
    plt.annotate(r'$ \alpha = $' + str(),(-0.75,6))
    plt.annotate(r'$ \beta = $' + str(),(-0.75,4.5))
    plt.xlabel('y')
    plt.ylabel(r'$\pi$')
    plt.axhline(0,color = 'black')
    plt.legend(frameon= False, fontsize="12")
    plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
    plt.ylim(-10,10)
    plt.xlim(-1,1)

    #plt.savefig('Fischer_pers1.pdf')

npoints = 1000

## Baseline
 = 11
 = 0.2
 = 0.95
plot_phil(,,,npoints)
plt.savefig('bl.pdf')
plt.show()

## Labor is elastic
 = 2
 = 0.2
 = 0.95
plot_phil(,,,npoints)
plt.savefig('low_phi.pdf')
plt.show()

## Price changes are more frequent
 = 11
 = 0.7
 = 0.95
plot_phil(,,,npoints)
plt.savefig('high_alpha.pdf')
plt.show()

## Agents become less patient
 = 11
 = 0.2
 = 0.8
plot_phil(,,,npoints)
plt.savefig('low_beta.pdf')
plt.show()
```

```python

## Some series for ideal prices, perfectly predictable
N = 1000
pstar = np.linspace(1,1000,N)

## Fischer contracts
p_fischer = pstar ##If ideal prices are perfectly predictable, Fischer firms hit them

## Taylor contracts
x_taylor = np.zeros(N)
p_taylor = np.zeros(N)
for t in range(N-1):
    x_taylor[t] = 0.5 * (pstar[t]+pstar[t+1]) 
    p_taylor[t] = 0.5*(x_taylor[t-1]+x_taylor[t])

plt.plot(p_fischer)
plt.plot(x_taylor)
plt.xlim(0,10)
plt.ylim(0,10)
```
